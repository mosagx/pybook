```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
(train_image, train_label), (test_image, test_label) = tf.keras.datasets.fashion_mnist.load_data()
```

```python
train_image.shape  # 60000 个 28*28 的元素
```

```python
train_label.shape
```

```python
test_image.shape, test_label.shape
```

```python
plt.imshow(train_image[0])
```

```python
np.max(train_image[0])
```

```python
train_label  # 训练结果集、期望
```

```python
train_image = train_image/255  # [0,1]区间
test_image = test_image/255    # [0,1]区间
```

```python
train_image.shape
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Flatten(input_shape=(28,28)))  # 28*28 像素图片
model.add(tf.keras.layers.Dense(128, activation='relu'))    # 输入128个隐藏单元，使用 relu 激活函数
model.add(tf.keras.layers.Dense(10, activation='softmax'))  # 输出10个概率值 [0, 9]，使用 softmax 激活函数
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['acc'])  # 编译模型
```

```python
model.fit(train_image, train_label, epochs=5)  # 训练模型
```

```python
model.evaluate(test_image, test_label) # 验证测试数据正确率
```

```python
# 转换为独热编码
train_label_onehot = tf.keras.utils.to_categorical(train_label)
```

```python
train_label_onehot[0]
```

```python
model.summary()
```

```python
# 训练结果集转换为独热类型
test_label_onehot = tf.keras.utils.to_categorical(test_label)
```

```python
test_label_onehot[0]
```

```python
# 独热训练模型
model_onehot = tf.keras.Sequential()
model_onehot.add(tf.keras.layers.Flatten(input_shape=(28, 28)))
model_onehot.add(tf.keras.layers.Dense(128, activation='relu'))
model_onehot.add(tf.keras.layers.Dense(10, activation='softmax'))
```

```python
# 独热编译
model_onehot.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])
```

```python
model_onehot.fit(train_image, train_label_onehot, epochs=5)
```

```python
predict = model_onehot.predict(test_image)
```

```python
predict.shape
```

```python
# 匹配各结果概率
predict[0]
```

```python
# 概率最大的为最终预测结果
np.argmax(predict[0])
```
